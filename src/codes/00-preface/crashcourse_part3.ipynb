# Crashkurs: Visualisieren mit Matplotlib

Matplotlib ist eine beliebte Alternative zu MATLAB, besonders in Kombination mit NumPy und SciPy zur Datenvisualisierung. Im Gegensatz zu MATLAB, das kostenpflichtig und Closed-Source ist, ist Matplotlib kostenlos, quelloffen und ermöglicht die Erstellung von Diagrammen in verschiedenen Formaten für Veröffentlichungen. Zudem bietet es eine steile Lernkurve, wodurch Nutzer schnell Fortschritte machen können.

# Einführung

Wir werden mit einem einfachen Graphen beginnen. So einfach, dass es nicht mehr einfacher geht.

Hinweis: Mit Hilfe des folgenden Befehls können wir das Format der Diagramme von `png` in `svg` ändern:

```python
%config InlineBackend.figure_formats = ['svg']
```

Ein Graph in Matplotlib ist eine zwei- oder dreidimensionale Darstellung, die mithilfe von Punkten, Kurven oder Balken Zusammenhänge visualisiert. Die x-Achse repräsentiert unabhängige Werte, die y-Achse abhängige Werte.

Wir verwenden das Untermodul `pyplot`, das eine prozedurale Schnittstelle zur objektorientierten Matplotlib-Bibliothek bietet. Die Befehle von `pyplot` ähneln MATLAB in Namen und Argumenten. Es ist üblich, `matplotlib.pyplot` als `plt` zu importieren. Im ersten Beispiel nutzen wir die `plot`-Funktion, die eine Liste von Werten als y-Werte interpretiert und die Indizes automatisch als x-Werte verwendet.


```python
import matplotlib.pyplot as plt

plt.plot([-1, -4.5, 16, 23, 15, 59])
plt.show()
```

Wir sehen einen zusammenhängenden Graphen, obwohl nur diskrete Werte für die Y-Achse (Ordinate) vorlagen. Für die X-Achse (Abszisse) wurden automatisch die Indizes 0, 1, 2, ... verwendet. 

Durch einen Formatstring beim Funktionsaufruf können wir einen Graphen mit diskreten Werten erzeugen, z. B. mit blauen Vollkreisen. Der Formatstring definiert dabei die Darstellung der Punkte:

```python
plt.plot([-1, -4.5, 16, 23, 15, 59], "o")
plt.show()
```

Wie einige vielleicht vermutet haben, kann man auch X-Werte an die Plot-Funktion übergeben. Im folgenden Beispiel nutzen wir eine Liste mit Vielfachen von 3 zwischen 0 und 21 als X-Werte:


```python
# die X-Werte:
days = list(range(0, 22, 3))
print(days)

# die Y-Werte:
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values)
plt.show()
```

Wir können das Aussehen unseres Graphen verbessern, indem wir die Achsen mit Bezeichnungen versehen. Dazu benutzen wir die ylabel- und xlabel-Funktionen von pyplot.

```python
days = list(range(1,9))
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values)
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.show()
```

Wir können eine beliebige Anzahl von (x, y, fmt)-Gruppen in einer Plot-Funktion spezifizieren. Im folgenden Beispiel benutzen wir zwei verschiedene Listen mit Y-Werten:


```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

plt.xlabel('Day')
plt.ylabel('Degrees Celsius')

plt.plot(days, celsius_min,  # Plottet verbundene Punkte von celcius_min
days, celsius_min, "oy",     # Plottet diskrete Punkte ("o") von celcius_min mit der Farbe Gelb ("y")
days, celsius_max,           # Plottet verbundene Punkte von celcius_max
days, celsius_max, "or")     # Plottet diskrete Punkte ("o") von celcius_min mit der Farbe Rot ("r")
plt.show()
```

Im folgenden Beispiel werden wir die NumPy-Funktion linspace verwenden. linspace
wird dazu benutzt, gleichmäßig verteilte Werte innerhalb eines spezifizierten Intervalls zu
erzeugen. Wir haben linspace ausführlich in unserem NumPy-Kapitel behandelt.


```python
import numpy as np

X = np.linspace(0, 2 * np.pi, 50, endpoint=True)
F = np.sin(X)

plt.plot(X, F)

startx, endx = -0.1, 2*np.pi + 0.1
starty, endy = -1.1, 1.1
plt.axis([startx, endx, starty, endy])
plt.show()
```

Das nächste Beispiel ist im Prinzip nichts Neues. Wir fügen lediglich zwei weitere Plots mit
diskreten Punkten hinzu:

```python
X = np.linspace(-2 * np.pi, 2 * np.pi, 50, endpoint=True)
F1 = 3 * np.sin(X)
F2 = np.sin(2*X)
F3 = 0.3 * np.sin(X)

startx, endx = -2 * np.pi - 0.1, 2*np.pi + 0.1
starty, endy = -3.1, 3.1
plt.axis([startx, endx, starty, endy])

plt.plot(X,F1)
plt.plot(X,F2)
plt.plot(X,F3)
plt.plot(X, F1, 'ro')
plt.plot(X, F2, 'bx')
plt.show()
```

Der Linienstil eines Plots kann durch die Parameter linestyle oder ls der plotFunktion beeinflusst werden. Sie können auf einen der folgenden Werte gesetzt werden:
’-’, ’–-’, ’-.’, ’:’, ’None’, ’ ’, ”
Wir können mit linewidth, wie der Name impliziert, die Linienstärke oder Liniendicke
setzen:

```python
X = np.linspace(0, 2 * np.pi, 50, endpoint=True)
F1 = 3 * np.sin(X)
F2 = np.sin(2*X)
F3 = 0.3 * np.sin(X)
F4 = np.cos(X)

plt.plot(X, F1, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, F2, color="red", linewidth=1.5, linestyle="--")
plt.plot(X, F3, color="green", linewidth=2, linestyle=":")
plt.plot(X, F4, color="grey", linewidth=2, linestyle="-.")
plt.show()
```

Mit der pyplot-Funktion `fill_between` ist es möglich, Flächen zwischen Kurven oder
Achsen zu schraffieren oder einzufärben. Im folgenden Beispiel füllen wir die Flächen zwischen der X-Achse und dem Graph der Funktion $sin(2X)$:

```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y, color='blue', alpha=1.00)
plt.fill_between(X, 0, Y, color='blue', alpha=.1) # Mit alpha wird die Transparenz beeinflusst
plt.show()
```

# Achsen- und Skalenteilung

Die Achsen eines Graphs werden in Matplotlib als "spines" bezeichnet. Wir zeigen, wie man diese an beliebige Stellen verschieben kann.

Die Funktion `plot` ist eine Methode der Klasse `Axes`. Normalerweise wird sie über `plt.plot` aufgerufen, wobei Matplotlib im Hintergrund eine `Axes`-Instanz erstellt. Bei mehreren Figuren wird es jedoch übersichtlicher, diese Instanzen selbst zu erzeugen, z. B. mit der Funktion `subplots`.

```python
fig, ax = plt.subplots()
```

Diese Funktion gibt zwei Werte zurück: `fig`, die gesamte Plotfläche, und `ax`, die Achse (Instanz der Klasse `Axes`), die den Bereich innerhalb des schwarzen Kastens bezeichnet.

Nun können wir `ax` verwenden, um Eigenschaften zu ändern. Mit `ax.spines` erhält man Zugriff auf die Achseneinstellungen. Wenn `ax` definiert ist, sollte die `plot`-Methode immer mit `ax.plot` aufgerufen werden.

```python
X = np.linspace(-2 * np.pi, 2 * np.pi, 70, endpoint=True)
F1 = np.sin(2* X)
F2 = (2*X**5 + 4*X**4 - 4.8*X**3 + 1.2*X**2 + X + 1)*np.exp(-X**2)

# Figur und Achse erzeugen
fig, ax = plt.subplots()

# Obere und rechte Achse unsichtbar werden lassen:
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Untere Achse auf die y=0 Position (in Datenkoordinaten) bewegen:
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

# Linke Achse auf die Position x == 0 (in Datenkoordinaten) bewegen:
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

ax.plot(X, F1)
ax.plot(X, F2)

plt.show()
```

Matplotlib hat bisher automatisch die Skalenbeschriftung ermittelt. Im letzten Beispiel war die X-Achse mit -8, -6, -4, -2, 0, 2, 4, 6, 8 und die Y-Achse mit -2.0, -1.5, -1.0, -0.5, 0, 0.5, 1.0, 1.5, 2.0 markiert. (Diese Werte können je nach Gerät variieren.)


Markierungen auf den Achsen nennt man "Ticks" und die zugehörigen Zahlen "Ticklabels". Um diese anzupassen, helfen die Methoden `set_xticks` und `set_xticklabels` von `Axes`. Mit `set_xticks` legt man die Positionen der Markierungen auf der x-Achse fest, indem man ein Array von Zahlen übergibt.

```python
fig, ax = plt.subplots()
xtick_pos = np.arange(10)
ax.set_xticks(xtick_pos)

print(xtick_pos)
```

Wie im obigen Beispiel zu sehen, werden die Ticklabels nach `set_xticks` angepasst. Daher sollte man immer zuerst `set_xticks` verwenden, um die Position der Markierungen festzulegen, und anschließend `set_xticklabels`, um die Labels zu ändern. Im folgenden Beispiel wird gezeigt, wie man die Ticks und Labels der x-Achse anpasst. Wichtig ist, dass die Listenlängen beider Funktionen übereinstimmen.

```python
fig, ax = plt.subplots()

xtick_pos = np.arange(4)
xtick_label = ['Berlin', 'London', 'Hamburg', 'Toronto']

ax.set_xticks(xtick_pos)
ax.set_xticklabels(xtick_label)

print(xtick_pos)
print(xtick_label)
```

Für die y-Achse gibt es natürlich auch die entsprechenden Methoden: `set_yticks` und `set_yticklabels`.

Gehen wir nochmal zurück zum vorigen Beispiel der Trigonometrischen Funktionen. Die
meisten werden eine Beschriftung der x-Achse in Teilen bzw. Vielfachen von pi bevorzugen.

```python
X = np.linspace(-2 * np.pi, 2 * np.pi, 70, endpoint=True)
X = np.linspace(-2 * np.pi, 2 * np.pi, 70, endpoint=True)
F1 = np.sin(X**2)
F2 = X * np.sin(X)

# Figur und Achse erzeugen
fig, ax = plt.subplots()

# der obere rechte Spine wird unsichtbar gemacht:
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# der untere Spine wird in Position y=0 gebracht:
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

# der linke Spine wird nach rechts zu x == 0 bewegt:
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# Die Ticks der x-Achse werden auf -2*pi, -pi, pi und 2*pi eingestellt
ax.set_xticks([-6.28, -3.14, 3.14, 6.28])
# Die Markierung werden angepasst
ax.set_xticklabels(['-2*pi', '-pi', 'pi', '2*pi'])

# Die Ticks der y-Achse werden angepasst
ax.set_yticks([-3, -1, 0, +1, 3])

# Plotten beider Funktionen
ax.plot(X, F1)
ax.plot(X, F2)

plt.show()
```

# Legenden hinzufügen

Um die Funktion eines Graphen leichter zu verstehen, sollten Legenden direkt im Diagramm erscheinen. Der Begriff "Legende" stammt aus dem Lateinischen und bedeutet „das, was zu lesen ist“. Ursprünglich auf Karten verwendet, erklären Legenden in Graphen die dargestellten Funktionen oder Werte. Mit der pyplot-Funktion `legend()` können Legenden einfach hinzugefügt werden, da sie die gezeichneten Linien speichert. Die Beschriftungen werden in der richtigen Reihenfolge als Liste übergeben. Für eine einzelne Linie ist die Reihenfolge unerheblich.

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])
ax.legend(['A simple line'])
plt.show()
```

Bei einem Plot mit mehrere Linien ist die Legende eigentlich schon unverzichtbar. Außerdem ist diese nicht nur auf Linien beschränkt.

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4]) # ansteigende Linie
ax.plot([4, 3, 2, 1]) # absteigende Linie
ax.plot([2, 4, 3, 1], 'o') # einige Punkte

ax.legend(['an ascending line', 'a descending line', 'some points'])
plt.show()
```

Hat man sich beim Plotten etwas mehr Mühe gemacht und den optionalen Argument `label` übergeben, kann man `legend` sogar ohne Argumente aufrufen.

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4], label='an ascending line') # ansteigende Linie
ax.plot([4, 3, 2, 1], label='a descending line') # absteigende Linie
ax.plot([2, 4, 3, 1], 'o', label='some points') # einige Punkte

ax.legend()
plt.show()
```

Matplotlib wählt automatisch die beste Position für die Legende. Möchte man dies selbst bestimmen, kann man den optionalen Parameter `loc` verwenden. Mögliche Werte sind: 'upper left', 'upper right', 'lower left', 'lower right', 'upper center', 'lower center', 'center left', 'center right', 'center' oder 'best'. Ohne Angabe wird standardmäßig 'best' verwendet.

```python
x = np.linspace(0, 25, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, '-b', label='sin')
plt.plot(x, y2, '-r', label='cos')
plt.legend(loc='upper left')
plt.ylim(-1.5, 2.0)
plt.show()
```
